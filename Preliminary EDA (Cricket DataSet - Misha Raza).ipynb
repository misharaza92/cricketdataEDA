```python
#The same bowler is for each over so we can estimate some of the missing bowlerIDs through over IDs.
#Facing BatsmanID cannot always be estimated with overID because the batsman may change if they take a run to the other side.
```

```python
# Fill missing values in 'bowlerId' using the mode of 'bowlerId' for each 'overHistoryId'
df3['bowlerId'] = df3.groupby('overHistoryId')['bowlerId'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else pd.NA))
```

```python
# For df3
columns_of_interest = ['overHistoryId', 'bowlerId']

# Filter rows where 'bowlerId' is null
null_bowler_rows = df3[df3['bowlerId'].isnull()]

# Display only the specified columns for the filtered rows
result = null_bowler_rows[columns_of_interest]
print(result)
```

```python
# Assuming df3 is your DataFrame with missing values in 'bowlerId'
# Assuming over_id_mapping is a DataFrame with 'overHistoryId' and 'bowlerId' columns

# Example over_id_mapping DataFrame
over_id_mapping = pd.DataFrame({
    'overHistoryId': ['over1', 'over2'],
    'bowlerId': ['bowlerA', 'bowlerB']
})

# Set 'overHistoryId' as the index for the mapping DataFrame
over_id_mapping.set_index('overHistoryId', inplace=True)

# Fill missing 'bowlerId' values based on the mapping
df3['bowlerId'] = df3['bowlerId'].fillna(df3['overHistoryId'].map(over_id_mapping['bowlerId']))

# Display the updated DataFrame
print(df3)
```

```python
#Boxplot  for Each Column with Prominent Outliers
for column in df1.columns:
    plt.figure(figsize=(12, 6))
    
    # Check if the data in the column is numeric
    if pd.api.types.is_numeric_dtype(df1[column]):
        # Convert boolean Series to numeric values
        numeric_values = df1[column].astype(float)
        
        # Boxplot with prominent outliers
        sns.boxplot(x=numeric_values, flierprops=dict(markerfacecolor='red', marker='D', markersize=8))
        plt.title(f'Boxplot for {column} in df1 with Prominent Outliers')
        plt.xlabel(column)
        plt.ylabel('Values')
        
        plt.show()
    else:
        # If the data is not numeric, print a message
        print(f"Skipping boxplot for {column} as the data is not numeric.")
```

```python
#Handling outliers
for column in df1.columns:
    # Check if the data in the column is numeric
    if pd.api.types.is_numeric_dtype(df1[column]):
        # Calculate IQR
        Q1 = df1[column].quantile(0.25)
        Q3 = df1[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Print column name
        print(f"Processing column: {column}")

        # Identify and remove outliers
        outliers = (df1[column] < lower_bound) | (df1[column] > upper_bound)
        df1 = df1[~outliers]
    else:
        print(f"Skipping outlier removal for {column} as the data is not numeric.")

# Display updated data without outliers
print(df1)
```

```python

#Checking out outliers after we have attempted to deal with them.
import pandas as pd

def calculate_z_scores_for_multiple_dfs(dataframes):
    outlier_counts_list = []

    for df in dataframes:
        z_scores = (df - df.mean(numeric_only=True)) / df.std(numeric_only=True)
        outliers = (z_scores > 2) | (z_scores < -2)
        outlier_counts = outliers.sum()
        outlier_counts_list.append(outlier_counts)

    return outlier_counts_list

# Assuming df1 to df7 are your DataFrames
df_list = [df1, df2, df3, df4, df5, df6, df7]
outlier_counts_list = calculate_z_scores_for_multiple_dfs(df_list)

for i, outlier_counts in enumerate(outlier_counts_list):
    print(f"Outliers count for df{i+1}:\n{outlier_counts}\n")
```
